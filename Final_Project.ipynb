```python
! pip install geopandas
! pip install geojson
import folium
import pandas as pd
import numpy as np
import geopandas as gpd
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display
```

```python
# load shedule from file saved to Github
url = 'https://raw.githubusercontent.com/Shiny-B/COCT_LoadShedding_Vis/main/Load_Shedding_All_Areas_Schedule.csv'
df = pd.read_csv(url, index_col=0)
df.reset_index(inplace=True)
df=df.rename(columns={"LoadsheddingStage":"Stage"})
#print(df.head())

#test if the filter works 
  #df2= pd.DataFrame.from_dict({'Stage': [1],'Time': ['19:00'],'Day':[22]})
  #print(df2.head())
  #Lup=pd.merge(df,df2,on=None, how ='inner')
  #print(Lup.head())
```

```python
from ipywidgets.widgets.widget_description import DescriptionWidget
# load geoJson areas from github
cpt_zones = "https://raw.githubusercontent.com/Shiny-B/COCT_LoadShedding_Vis/main/Loadshedding_Blocks.geojson"

# create a colourmap to load onto our final graphic
from branca.colormap import linear
colormap = linear.YlOrBr_09.scale(0,1)
#print(colormap(1.0))
#colormap

# create function to add filtered overlay onto the map
def f(stage,day,time):
  # Read in the table and find if the zone is in loadshedding
  df2= pd.DataFrame.from_dict({'Stage': stage,'Time': time,'Day':[day]})
  Lup=pd.merge(df,df2,on=None, how ='inner')
  Lup= Lup.drop(columns=['Stage', 'Day'])
  Lup=Lup.transpose()
  Lup.reset_index(inplace=True)
  Lup=Lup.rename(columns={0:"Loadshed",'index':'BlockID'})
  print(Lup.head(15))
  zone_dict = Lup.set_index("BlockID")["Loadshed"]
  print(zone_dict["9"])
  folium.GeoJson(
    cpt_zones,
    name='CPT Loadshedding Zones',
    style_function=lambda feature: {
        "fillColor": colormap(zone_dict[feature["properties"]["BlockID"]]),
        "color": "black",
        "weight": 1,
        "dashArray": "5, 5",
        "fillOpacity": 0.3,
        },
    ).add_to(m)
  folium.LayerControl().add_to(m)
  #return

#create a map
m = folium.Map(location=[-33.932258, 18.495384],
               zoom_start=10,
               tiles="cartodbpositron",
               )

# add interactive buttons that will be used for creating maps
caption1=widgets.Label(value='This program will display a map with the all the loadshedding areas for cape town to help plan your day is you are travelling between zones. Highlighted in yellow where there is meant to be power and red when the zone is experiencing loadshedding. This map is for the city of cape town only so does not show the Eskom fed areas.')
caption2=widgets.Label(value='To run this interactive map the loadshedding stage and the date and time must be set. Unfortunately i have not been able to update the map dynamically so the map can only load once:( .The following steps are required')
caption3=widgets.Label(value='1) move the Stage Slider to the correct load sheeding stage you want to check ')
caption4=widgets.Label(value='2) select the day of the month and time in the Day & Time interactive fields')
caption5=widgets.Label(value='3) select the Run Interact button to generate the overlay')
caption5=widgets.Label(value='4) select the Display Map button to generate the overlay')
display(caption1, caption2, caption3, caption4, caption5 )
im= interact_manual(f,
                    #button = widgets.Button(description="Click Me!") ,                  
                    stage=widgets.IntSlider(
                        value=2,
                        min=1,
                        max=8,
                        Step=1,
                        description='Stage '
                        ),
                    day=widgets.IntSlider(
                        value=17,
                        min=1,
                        max=31,
                        Step=1,
                        description='Day'
                        ),
                    time=['00:00',
                          '01:00',
                          '02:00',
                          '03:00',
                          '04:00',
                          '05:00',
                          '06:00',
                          '07:00',
                          '08:00',
                          '09:00',
                          '10:00',
                          '11:00',
                          '12:00',
                          '13:00',
                          '14:00',
                          '15:00',
                          '16:00',
                          '17:00',
                          '18:00',
                          '19:00',
                          '20:00',
                          '21:00',
                          '22:00',
                          '23:00',
                          '24:00']
                    );


# A button to display the map
but1 = widgets.Button(description="Display Map")
output = widgets.Output()

display(but1, output)

def on_button_clicked(b):
  display(m)
  print("Button clicked.")

but1.on_click(on_button_clicked)


# A button to create a new map
but2 = widgets.Button(description="Hide Map")
output = widgets.Output()

display(but2, output)

def on_button_clicked2(b):
  hide(m)
  print("Button clicked.")

but2.on_click(on_button_clicked2)
```
